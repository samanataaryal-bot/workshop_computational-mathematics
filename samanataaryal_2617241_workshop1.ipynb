a. Write a function fib (n) that prompts the number of terms “n” from the user anddisplays first “n” terms of Fibonacci sequence iteratively.

```python
def fib(n):
  a=0
  b=1
  if n == 1:
    print(a)
  else:
    print(a)
    print(b)
    for i in range(2,n):
      c=a+b
      print(c)
      a=b
      b=c
n=int(input("Enter number of terms: "))
fib(n)
```

b. Write a function myPow(x, n) that returns 𝑥𝑛, where and n is a non-negative integer.Do not use the ** operator or the math.pow function - use one while loop.

```python
def myPow(x, n):
  prod = 1
  for i in range(n):
    prod *= x
  return prod

print(myPow(2, 3))
```

c. Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

```python
def SumDigits(n):
  sum = 0
  while n!=0:
    sum += n%10
    n //=10
  return sum

print(SumDigits(234))
```

d. Write a function squaredSum that returns the sum of the squares of an array/list ofintegers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Pythonto show the logic in your solution, do not use any library functions.

```python
def squaredSum(nums):
  sum = 0
  for i in range(len(nums)):
    nums[i] = nums[i] * nums[i]
    sum += nums[i]
  return sum
print (squaredSum([2,3,4]))
```

e. Write a function called sumDigits that is given an integer num and returns the sum ofthe digits of num. Eg. sumDigits(6) gives 21. Do not use any libraryfunctions (ifthere is one).

```python
def sumDigits(n):
  sum = 0
  while n!=0:
    sum += n
    n -= 1
  return sum

print(sumDigits(6))
```
